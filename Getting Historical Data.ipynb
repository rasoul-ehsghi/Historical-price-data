```python
import yfinance as yf

#Define ticker symbol

ticker_symbol = "AAPL"

# Create a ticker object

ticker = yf.Ticker(ticker_symbol)

# Download Historical Data
historical_data = ticker.history(start = "1990-01-01", end="2022-12-21", interval='1d')
historical_data
```

```python
# inCase if you want to download the last days or shorter period use the following method:

history_data = ticker.history(period = "20d", interval = "1d")
history_data
```
